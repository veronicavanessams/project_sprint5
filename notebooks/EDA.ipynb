```python
import pandas as pd
df = pd.read_csv('../vehicles.csv')
df['odometer'] = pd.to_numeric(df['odometer'], errors='coerce')
df = df.dropna(subset=['odometer'])
df.head()
```

```python
import plotly.express as px
import plotly.io as pio

fig = px.histogram(df, x='odometer', title='Distribuição do Hodômetro')

# salva um arquivo HTML independente (não precisa servidor local)
pio.write_html(fig, file='hist_odometer.html', auto_open=True, include_plotlyjs='cdn')
```

```python
# 1) HISTOGRAMA — já funcionou e salva HTML
import pandas as pd, plotly.express as px, plotly.io as pio
df = pd.read_csv('../vehicles.csv')
df['odometer'] = pd.to_numeric(df['odometer'], errors='coerce')
df = df.dropna(subset=['odometer'])
fig1 = px.histogram(df, x='odometer', title='Distribuição do Hodômetro')
pio.write_html(fig1, file='hist_odometer.html', auto_open=False, include_plotlyjs='cdn')
```

```python
# 2) SCATTER simples e também salvo
base = df.dropna(subset=['model_year', 'price'])
fig2 = px.scatter(base, x='model_year', y='price',
                  title='Preço vs Ano do Modelo', opacity=0.4)
pio.write_html(fig2, file='scatter_price_year.html', auto_open=False, include_plotlyjs='cdn')
```
